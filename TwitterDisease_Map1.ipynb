```python
def data_To_diction(disease,country):
    
    print(country)
    print(disease)
    if country not in Countryarray:
        Countryarray.append(country)
        print(Countryarray)
        diction={}
        map_dictionary.update({country:diction}) 
        print(map_dictionary)
    map_dictionary.get(country).update({disease:map_dictionary.get(country).get(disease)+1 if disease  in map_dictionary.get(country) else 1 })
    print(map_dictionary)
    return True
```
